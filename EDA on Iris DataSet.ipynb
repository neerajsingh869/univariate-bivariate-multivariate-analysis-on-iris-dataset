### Univariate Analysis

```python
data['species'].unique()
```

```python
data_setosa = data.loc[data['species']=='setosa']
data_versicolor = data.loc[data['species']=='versicolor']
data_virginica = data.loc[data['species']=='virginica']
```

```python
data_setosa.shape
```

```python
# Let's plot now
plt.figure(figsize = (8, 4))
plt.plot(data_setosa['sepal_length'], np.zeros_like(data_setosa['sepal_length']), '*', markersize = 12)
plt.plot(data_versicolor['sepal_length'], np.zeros_like(data_versicolor['sepal_length']), 'o',markersize = 12)
plt.plot(data_virginica['sepal_length'], np.zeros_like(data_virginica['sepal_length']), '^',markersize = 12)
plt.show()
```
